Ce notebook contient l'Arbre Binaire de Recherche (ABR) ; les parcours prefixé, infixé et postfixé ; les méthodes existe et éliminer

```python
class Noeud: #Création de la classe Noeud
  def __init__(self,val=None):
    self.val = val
    self.gauche = None
    self.droite = None

  def __str__(self):
    return str(self.val)

class ABR: #Création de la classe ABR
  def __init__(self,val=None):
    if val is not None:
      self.racine = Noeud(val)
    else:
      self.racine = None

  def ajouter(self,val):
    if self.racine is None:
      self.racine = Noeud(val)
      return

    n = self.racine

    while (n is not None):
      if val <= n.val:
        if n.gauche is None:
          n.gauche = Noeud(val)
          return
        n = n.gauche

      else:
        if n.droite is None:
          n.droite = Noeud(val)
          return
        n = n.droite

  def parcoursInfixCroissant(self): #Création du parcours infix croissant
    self.parcoursInfixCroissantR(self.racine)

  def parcoursInfixCroissantR(self,noeud):
    if noeud is None:
      return
    self.parcoursInfixCroissantR(noeud.gauche)
    print(noeud.val)
    self.parcoursInfixCroissantR(noeud.droite)

  def parcoursInfixDecroissant(self): #Création du parcours infix décroissant
    self.parcoursInfixDecroissantR(self.racine)

  def parcoursInfixDecroissantR(self,noeud):
    if noeud is None:
      return
    self.parcoursInfixDecroissantR(noeud.droite)
    print(noeud.val)
    self.parcoursInfixDecroissantR(noeud.gauche)

  def parcoursPrefixe1(self): #Création du parcours préfixe (de gauche à droite)
    self.parcoursPrefixeR1(self.racine)

  def parcoursPrefixeR1(self,noeud):
    if noeud is None:
      return
    print(noeud.val)
    self.parcoursPrefixeR1(noeud.gauche)
    self.parcoursPrefixeR1(noeud.droite)

  def parcoursPrefixe2(self): #Création du parcours préfixe (de droite à gauche)
    self.parcoursPrefixeR2(self.racine)

  def parcoursPrefixeR2(self,noeud):
    if noeud is None:
      return
    print(noeud.val)
    self.parcoursPrefixeR2(noeud.droite)
    self.parcoursPrefixeR2(noeud.gauche)

  def parcoursPostfixe1(self): #Création du parcours postfixe (par génération de gauche à droite)
    self.parcoursPostfixeR1(self.racine)

  def parcoursPostfixeR1(self,noeud):
    if noeud is None:
      return
    self.parcoursPostfixeR1(noeud.gauche)
    self.parcoursPostfixeR1(noeud.droite)
    print(noeud.val)

  def parcoursPostfixe2(self): #Création du parcours postfixe (par génération de droite à gauche)
    self.parcoursPostfixeR2(self.racine)

  def parcoursPostfixeR2(self,noeud):
    if noeud is None:
      return
    self.parcoursPostfixeR2(noeud.droite)
    self.parcoursPostfixeR2(noeud.gauche)
    print(noeud.val)

  def existe(self,val): #Méthode pour vérifier l'existence d'un noeud ou racine
    if (self.racine is not None):
      return self.existeR(self.racine, val)
    return False

  def existeR(self,noeud,val):
    if (noeud.val==val):
      return True
    if val < noeud.val:
      if noeud.gauche is not None:
        return self.existeR(noeud.gauche,val)
      else:
        return False
    else:
      if noeud.droite is not None:
        return self.existeR(noeud.droite,val)
      else:
        return False

  def successeur(self, noeud): #Méthode pour éliminer un noeud
    courant=noeud
    while(courant.gauche is not None):
      courant=courant.gauche
    return courant

  def predecesseur(self,noeud):
    courant=noeud
    while(courant.droite is not None):
      courant=courant.droite
    return courant

  def eliminer(self,val):
    return self.eliminerR(self.racine,val)

  def eliminerR(self,noeud,val):
    if noeud is None:
      return None

    elif (val < noeud.val):
      noeud.gauche=self.eliminerR(noeud.gauche,val)

    elif(val > noeud.val):
      noeud.droite=self.eliminerR(noeud.droite,val)

    else:
      if noeud.gauche is None:
        temp=noeud.droite
        noeud=None
        return temp

      elif noeud.droite is None:
        temp=noeud
        noeud=None
        return temp

      succ=self.successeur(noeud.droite)

      pred=self.predecesseur(noeud.gauche)

      noeud.val=succ.val

      noeud.droite=self.eliminerR(noeud.droite,succ.val)

      noeud.gauche=self.eliminerR(noeud.gauche,succ.val)

    return noeud

#Création de la racine de l'arbre
abr = ABR(10)
print(abr.racine.val)

#Création des feuilles de l'arbre
abr.ajouter(5)
print(abr.racine.gauche.val)

abr.ajouter(1)
print(abr.racine.gauche.gauche.val)

abr.ajouter(6)
print(abr.racine.gauche.droite.val)

abr.ajouter(9)
print(abr.racine.gauche.droite.droite.val)

abr.ajouter(11)
print(abr.racine.droite.val)

#Méthode de parcours de l'arbre !
print("Parcours Infix Croissant :")
abr.parcoursInfixCroissant()

print("Parcours Infix Décroissant :")
abr.parcoursInfixDecroissant()

print("Parcours Prefixe (de gauche à droite):")
abr.parcoursPrefixe1()

print("Parcours Prefixe (de droite à gauche):")
abr.parcoursPrefixe2()

print("Parcours Postfixe (par génération de gauche à droite):")
abr.parcoursPostfixe1()

print("Parcours Postfixe (par génération de droite à gauche):")
abr.parcoursPostfixe2()

abr.existe(20) #Tester la méthode existe avec un nombre

racine=abr.eliminer(1) #Tester la méthode éliminer avec un nombre
print("Liste avec le noeud supprimé :")
abr.parcoursInfixCroissant()
```
